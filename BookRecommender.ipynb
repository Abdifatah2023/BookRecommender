```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
listfile = os.listdir(path)
listfile
```

```python
df = pd.read_csv(path + "/" + "books.csv")
df.head(5)
```

```python
missing_vals = df.isna().sum()
missing_vals
```

```python
df['p_year_na'] = df['published_year'].isna()
df
```

```python
df['published_year'] = df['published_year'].fillna('NA')
df['published_year']
```

```python
df.drop(columns=['p_year_na'], inplace=True)
```

```python
missing = int(df['average_rating'].isna().sum())
missing
```

```python
df['av_rating_na'] = df['average_rating'].isna()
df
```

```python
df['average_rating'] = df['average_rating'].fillna(df['average_rating'].mean())
```

```python
df.drop(columns=['av_rating_na'], inplace=True)
df
```

```python
df['missing_num_pages'] = df['num_pages'].isna()
df
```

```python
df['num_pages'] = df['num_pages'].fillna('NA')
```

```python
df.drop(columns=['missing_num_pages'], inplace=True)
df
```

```python
df = df[df['ratings_count'] >= 10]
```

```python
missingVals = df.isna().sum()
missingVals
```

```python
df['missing_sbt'] = df['subtitle'].isna()
df['missing_author'] = df['authors'].isna()
df['missing_ctgry'] = df['categories'].isna()
df['mis_thumbnail'] = df['thumbnail'].isna()
df['mis_description'] = df['description'].isna()
df
```

```python
df['subtitle'] = df['subtitle'].fillna('No Subtitle')
df['authors'] = df['authors'].fillna('Unknown Author')
df['categories'] = df['categories'].fillna('Uncategorized')
df['thumbnail'] = df['thumbnail'].fillna('No Image')
df['description'] = df['description'].fillna('No Description')
```

```python
df.drop(columns=['missing_sbt'], inplace=True)
df.drop(columns=['missing_author'], inplace=True)
df.drop(columns=['missing_ctgry'], inplace=True)
df.drop(columns=['mis_thumbnail'], inplace=True)
df.drop(columns=['mis_description'], inplace=True)

df
```

```python
df = df[df['description'].str.lower() != 'no description']
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.drop(columns=['isbn10'], inplace=True)
df
```

```python
df['description_len'] = df['description'].str.split().str.len()
df
```

```python
df.loc[df['description_len'].between(15, 25), 'description']  # finding the cut off description len
```

```python
df_updated = df[df['description_len'] >= 25]
df_updated
```

```python
df_updated.drop(columns=['description_len'], inplace=True)
df_updated
```

```python
df_updated['title_&_subtitle'] = (
    np.where(df_updated['subtitle']=='No Subtitle', df_updated['title'],
             df_updated[['title', 'subtitle']].astype(str).agg(': '.join, axis=1)))

df_updated
```

```python
df_updated.drop(columns=['title', 'subtitle'], inplace=True)
df_updated
```

```python
df_updated["tagged_description"] = df_updated[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
df_updated
```

```python
# categories
df_updated['categories'].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
df_updated.to_csv('updated_books.csv', index=False)
```
